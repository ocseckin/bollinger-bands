```python
def bollinger(research, start_year,start_month,start_day):
    "it lets you see bollinger bands w/ 20 moving averages"
    import pandas as pd
    import matplotlib.pyplot as plt
    import pandas_datareader.data as web
    from datetime import datetime
    from datetime import timedelta
    start_date = datetime(start_year,start_month,start_day) - timedelta(days=20)
    
    stock = web.DataReader(research+'.IS','yahoo',start=start_date)
    
    fig,ax = plt.subplots(figsize=(14,7))
    ax.plot(stock['Adj Close'][19:],c='blue',label='Adj Close')
    ax.plot(stock['Adj Close'].rolling(20).mean() + 2*stock['Adj Close'].rolling(20).std(),c='grey',label='Upper Band')
    ax.plot(stock['Adj Close'].rolling(20).mean() - 2*stock['Adj Close'].rolling(20).std(),c='grey', label ='Lower Band')
    ax.legend(frameon=False)
    
    ax2 = ax.twinx()
    ax2.plot(stock['Volume'][19:],alpha=0.3,ls='--')
```

```python
bollinger('VAKKO',2019,1,1)
```
